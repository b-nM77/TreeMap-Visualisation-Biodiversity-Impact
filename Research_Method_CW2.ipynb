```python
#Treemap : Biodiversity Impact by Diet, Age Group, and Gender
#Import Libraries
import pandas as pd
import plotly.express as px
```

```python
# Load the dataset
data = pd.read_csv('Results_21Mar2022.csv')
```

```python
# Check for missing values
print("Checking for missing values:")
print(data.isnull().sum())
#print("\n")

# Check data types
print("Checking data types:")
print(data.dtypes)
#print("\n")
```

```python
# Group the data
grouped_data = data.groupby(['diet_group', 'age_group', 'sex'], as_index=False).agg({
    'mean_bio': 'mean',
    'mean_watuse': 'mean'
})

# Create the treemap
fig = px.treemap(
    grouped_data,
    path=['diet_group', 'age_group', 'sex'],
    values='mean_bio',
    color='mean_watuse',
    color_continuous_scale=['#f5f5dc', '#deb887', '#8b4513'],  # Cream to brown colors
    title='Visualization : Biodiversity Impact by Diet, Age Group, and Gender'
)

# Layout details
fig.update_layout(margin=dict(t=50, l=25, r=25, b=25))

# Display the figure
fig.show()
```
